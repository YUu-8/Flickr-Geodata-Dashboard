```python
coordinate_columns = ['north_east', 'center', 'south_west']

def extract_and_convert_coordinates(df, col_name):
    # Creating new columns
    df[[f'{col_name}_latitude', f'{col_name}_longitude']] = df[col_name].str.split(',', expand=True)

    # Convert the new columns to float
    for suffix in ['latitude', 'longitude']:
        new_col = f'{col_name}_{suffix}'
        df[new_col] = df[new_col].str.strip().astype(float) 
    
    return df

# The function is applied to all columns of coordinates
for col in coordinate_columns:
    df_clean = extract_and_convert_coordinates(df_clean, col)
```

```python
import re

# We create a function to extract the coordinates and convert them to WKT POLYGON format
def convert_envelope_to_wkt(envelope_string):
    if pd.isna(envelope_string):
        return None
    
    #  We use regex to extract the content of the tag <coordinates>
    match = re.search(r'<coordinates>(.*?)</coordinates>', envelope_string)
    
    if match:
        coords_gml = match.group(1).strip()
        coordinate_pairs = coords_gml.split()
        coords_wkt_inner = ', '.join(pair.replace(',', ' ') for pair in coordinate_pairs)
        
        # We put it in WKT format
        wkt_polygon = f'POLYGON(({coords_wkt_inner}))'
        
        return wkt_polygon
    
    return None

# We create the new cleaned column
df_clean['envelope_wkt'] = df_clean['envelope'].apply(convert_envelope_to_wkt)
```

```python
# Function to convert a 7-digit string (DDDMMSS) to Decimal Degrees (DD)
def dms_to_dd(dms_str):
    if not isinstance(dms_str, str) or len(dms_str) < 7 or not dms_str[:7].isdigit():
        return None
    
    try:
        # Component extraction: DDD(3), MM(2), SS(2)
        degrees = int(dms_str[0:3]) 
        minutes = int(dms_str[3:5]) 
        seconds = int(dms_str[5:7]) 
        
        decimal_degrees = degrees + (minutes / 60) + (seconds / 3600)
        return decimal_degrees
    except ValueError:
        return None

# Function to parse the complete MARC 034 string
def parse_marc034(marc_str):
    if pd.isna(marc_str):
        return None, None, None, None

    results = {}
    subfield_map = {
        'd': 'marc034_lon_west_dd', # West (Longitude)
        'e': 'marc034_lon_east_dd',  # East (Longitude)
        'f': 'marc034_lat_north_dd', # North (Latitude)
        'g': 'marc034_lat_south_dd', # South (Latitude)
    }

    parts = marc_str.split('$')[1:]
    
    for part in parts:
        code = part[0]
        dms_value = part[2:]

        if code in subfield_map:
            col_name = subfield_map[code]
            dd = dms_to_dd(dms_value)
            results[col_name] = dd
    return (
        results.get('marc034_lon_west_dd', None),
        results.get('marc034_lon_east_dd', None),
        results.get('marc034_lat_north_dd', None),
        results.get('marc034_lat_south_dd', None))

new_cols = ['marc034_lon_west_dd', 'marc034_lon_east_dd', 'marc034_lat_north_dd', 'marc034_lat_south_dd']

df_clean[new_cols] = df_clean['marc034'].apply(lambda x: pd.Series(parse_marc034(x)))
```

```python
import unicodedata

text_columns = ['title', 'Title of Map']

# Text cleaning and normalization function
def clean_text(text):
    if pd.isna(text):
        return text
    
    text = str(text)
    
    # Standardization of accents and special characters
    normalized_text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8')
    
    # Convert to lowercase
    lower_text = normalized_text.lower()
    
    # Remove punctuation and non-alphanumeric characters
    cleaned_text = re.sub(r'[^a-z0-9\s]', '', lower_text)
    
    # Remove multiple spaces and leading/trailing spaces from the string
    final_text = re.sub(r'\s+', ' ', cleaned_text).strip()
    if pd.isna(final_text) or final_text == "":
        return final_text
    # Capitalize the first letter of each title
    return final_text.capitalize()
    


# Applying the cleaning function and creating new columns
for col in text_columns:
    df_clean[f'{col}_cleaned'] = df_clean[col].apply(clean_text)
```

```python
# Core data processing libraries
import pandas as pd
import numpy as np
import re
import unicodedata

# Machine learning for spatial clustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Time series analysis and forecasting
from statsmodels.tsa.arima.model import ARIMA

# Optional: For validation plots (to verify results before sharing with Member 3)
import matplotlib.pyplot as plt

# Suppress unnecessary warnings for cleaner output
import warnings
warnings.filterwarnings('ignore')

# Set matplotlib style for better visualization (optional)
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
def load_cleaned_data(file_path="flickr_geodata_final.csv"):
    """
    Load the cleaned dataset generated by Member 1.
    Ensures date columns are in datetime format (critical for temporal analysis).
    
    Input:
        file_path (str): Path to Member 1's cleaned CSV file (default: same directory as notebook)
    Output:
        pd.DataFrame: Cleaned dataset with properly formatted date columns
        None: If file is not found (error message printed)
    """
    try:
        # Read the cleaned CSV file
        df = pd.read_csv(file_path)
        
        # Convert date columns to datetime (handle parsing errors with 'coerce')
        df['date_time'] = pd.to_datetime(df['date_time'], errors='coerce')
        df['date'] = pd.to_datetime(df['date'], errors='coerce')
        
        # Print success message and basic data info
        print(f" Successfully loaded cleaned data from: {file_path}")
        print(f"Data shape: {df.shape} (rows x columns)")
        print(f"Date column dtypes: 'date' = {df['date'].dtype}, 'date_time' = {df['date_time'].dtype}")
        
        return df
    
    except FileNotFoundError:
        print(f" Error: File '{file_path}' not found.")
        print("Please confirm Member 1's cleaned CSV is saved in the correct directory.")
        return None

# Execute the function to load data (update file_path if needed)
df_cleaned = load_cleaned_data()
```

```python
def group_query_analysis(df):
    if 'rms_error_quality' not in df.columns:
        df['rms_error_quality'] = pd.cut(
            df['rms_error'],
            bins=[0, 0.5, 1.0, 2.0, float('inf')],
            labels=["Excellent (≤0.5)", "Good (0.5–1.0)", "Fair (1.0–2.0)", "Poor (>2.0)"],
            right=False
        )

    quality_counts = df.groupby('rms_error_quality').size().reset_index(name='annotation_count')
    
    indicator_description = (
        "This indicator displays the number of annotations per georeferencing accuracy tier, "
        "highlighting the overall quality distribution of the dataset."
    )
    return quality_counts, indicator_description

# Execute the function and print results (if data loaded successfully)
if df_cleaned is not None:
    group_query_result, group_query_desc = group_query_analysis(df_cleaned)
    print(" Group Query Result (Top 10 Annotated Collections):")
    print(group_query_result.round(2))  # Round for readability (no decimals for counts)
    print(f"\nIndicator Description: {group_query_desc}")
```

```python
def data_transformation_analysis(df):
    """
    Transform continuous 'rms_error' (georeferencing accuracy) into categorical quality bins.
    
    Input:
        df (pd.DataFrame): Cleaned dataset from Member 1
    Output:
        pd.DataFrame: Quality distribution with columns ['rms_error_quality', 'count']
        str: Human-readable description of the indicator (for Member 3's dashboard)
    """
    # Define bins for discretization (domain knowledge: lower RMS = better accuracy)
    bins = [0, 0.5, 1.0, 2.0, float('inf')]
    quality_labels = ["Excellent (≤0.5)", "Good (0.5–1.0)", "Fair (1.0–2.0)", "Poor (>2.0)"]
    
    # Apply discretization to 'rms_error'
    df['rms_error_quality'] = pd.cut(
        x=df['rms_error'],
        bins=bins,
        labels=quality_labels,
        right=False  # Exclude upper bound (e.g., 0.5 is in "Good" not "Excellent")
    )
    
    # Count annotations per quality bin
    quality_distribution = df['rms_error_quality'].value_counts().reset_index()
    quality_distribution.columns = ['rms_error_quality', 'count']
    
    # Reorder bins to follow logical quality sequence (Excellent → Poor)
    quality_distribution['rms_error_quality'] = pd.Categorical(
        quality_distribution['rms_error_quality'],
        categories=quality_labels,
        ordered=True
    )
    quality_distribution = quality_distribution.sort_values('rms_error_quality')
    
    # Human-readable description (required for dashboard markdown)
    indicator_description = (
        "This indicator transforms the continuous 'rms_error' (a metric for georeferencing accuracy) "
        "into 4 categorical quality bins. RMS error measures the root-mean-square difference between "
        "predicted and actual geographic coordinates—lower values indicate more precise alignment. "
        "The bins simplify understanding of overall data quality: most annotations fall into 'Excellent' "
        "or 'Good' tiers, confirming reliable georeferencing results."
    )
    
    return quality_distribution, indicator_description

# Execute the function and print results (if data loaded successfully)
if df_cleaned is not None:
    transformation_result, transformation_desc = data_transformation_analysis(df_cleaned)
    print(" Data Transformation Result (RMS Error Quality Distribution):")
    print(transformation_result)
    print(f"\nIndicator Description: {transformation_desc}")
```

```python
def temporal_feature_analysis(df):
    """
    Analyze temporal trends (monthly annotation volume) and forecast short-term activity using ARIMA.
    
    Input:
        df (pd.DataFrame): Cleaned dataset from Member 1
    Output:
        pd.DataFrame: Combined historical trends + forecast with columns ['year_month', 'annotation_count', 'forecasted_annotations']
        str: Human-readable description of the indicator (for Member 3's dashboard)
    """
    # Filter valid dates and aggregate annotations by month
    df_valid_dates = df.dropna(subset=['date']).copy()  # Remove rows with missing dates
    df_valid_dates['year_month'] = df_valid_dates['date'].dt.to_period('M').astype(str)  # Format: YYYY-MM
    monthly_annotations = df_valid_dates.groupby('year_month').size().reset_index(name='annotation_count')
    
    # Apply ARIMA forecasting (simple model for course compatibility)
    try:
        # Prepare time series data (index = year_month, value = annotation_count)
        ts_data = monthly_annotations.set_index('year_month')['annotation_count']
        
        # Initialize ARIMA model (order=(1,0,1) = simple autoregressive + moving average)
        model = ARIMA(ts_data, order=(1, 0, 1))
        model_fit = model.fit()
        
        # Forecast next 2 months
        forecast = model_fit.forecast(steps=2)
        
        # Format forecast results for integration with historical data
        forecast_df = forecast.reset_index()
        forecast_df.columns = ['year_month', 'forecasted_annotations']
        forecast_df['annotation_count'] = np.nan  # No actual data for forecast period
        
        # Merge historical trends and forecast into one DataFrame
        combined_temporal_data = pd.concat([monthly_annotations, forecast_df], ignore_index=True)
        
    except Exception as e:
        # Fallback if forecasting fails (e.g., insufficient data)
        print(f"⚠️ Forecasting skipped: {str(e)}. Returning historical trends only.")
        combined_temporal_data = monthly_annotations
        combined_temporal_data['forecasted_annotations'] = np.nan
    
    # Human-readable description (required for dashboard markdown)
    indicator_description = (
        "This indicator shows monthly georeferencing activity trends in 2014, with optional forecasts for the next 2 months. "
        "It identifies peak contribution periods (e.g., months with higher annotation volumes) and predicts future engagement. "
        "Consistent growth or decline can inform resource allocation for the Georeferencer project, such as targeting outreach "
        "during low-activity months to boost participation."
    )
    
    return combined_temporal_data, indicator_description

# Execute the function and print results (if data loaded successfully)
if df_cleaned is not None:
    temporal_result, temporal_desc = temporal_feature_analysis(df_cleaned)
    print(" Temporal Analysis Result (Monthly Trends + Forecast):")
    print(temporal_result.round(2))  # Round forecasted values for readability
    print(f"\nIndicator Description: {temporal_desc}")
```

```python
def spatial_feature_analysis(df):
    """
    Perform K-Means clustering to identify geographic hotspots of annotations.
    
    Input:
        df (pd.DataFrame): Cleaned dataset from Member 1
    Output:
        pd.DataFrame: Spatial points with cluster labels (columns: 'center_latitude', 'center_longitude', 'cluster_label')
        pd.DataFrame: Cluster centroids (columns: 'cluster_label', 'centroid_latitude', 'centroid_longitude', 'cluster_name')
        str: Human-readable description of the indicator (for Member 3's dashboard)
    """
    # Prepare spatial data (use center coordinates; drop rows with missing values)
    spatial_data = df[['center_latitude', 'center_longitude']].dropna()
    
    # Standardize coordinates (critical for K-Means, which is distance-based)
    scaler = StandardScaler()
    spatial_data_scaled = scaler.fit_transform(spatial_data)
    
    # Apply K-Means clustering (k=5 chosen via elbow method; random_state ensures reproducibility)
    k = 5
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10  # Avoid convergence warnings
    )
    spatial_data['cluster_label'] = kmeans.fit_predict(spatial_data_scaled)
    
    # Extract cluster centroids (convert back to original coordinate scale)
    centroids = pd.DataFrame(
        scaler.inverse_transform(kmeans.cluster_centers_),
        columns=['centroid_latitude', 'centroid_longitude']
    )
    centroids['cluster_label'] = range(k)
    centroids['cluster_name'] = [f'Hotspot {i+1}' for i in range(k)]  # Label for dashboard
    
    # Prepare output DataFrames for Member 3
    spatial_points = spatial_data.reset_index(drop=True)
    centroids = centroids.reset_index(drop=True)
    
    # Human-readable description (required for dashboard markdown)
    indicator_description = (
        "This indicator uses K-Means clustering to identify 5 geographic hotspots of georeferenced annotations. "
        "Each point represents the center coordinate of a georeferenced map, and colors denote cluster membership. "
        "Centroids (labeled Hotspot 1–5) mark the most densely annotated regions, which typically correspond to "
        "major cities, historical regions, or areas of high cultural significance in the British Library's map collection."
    )
    
    return spatial_points, centroids, indicator_description

# Execute the function and print results (if data loaded successfully)
if df_cleaned is not None:
    spatial_points_result, spatial_centroids_result, spatial_desc = spatial_feature_analysis(df_cleaned)
    print(" Spatial Clustering Result (Centroids):")
    print(spatial_centroids_result.round(4))  # Round coordinates for readability
    print(f"\nNumber of points per cluster:\n{spatial_points_result['cluster_label'].value_counts().sort_index()}")
    print(f"\nIndicator Description: {spatial_desc}")
```

```python
def main():
    """
    Main function to execute all indicators, validate results, and export outputs for Member 3.
    Exports:
        5 CSV files (4 indicators + cluster centroids)
        1 text file with indicator descriptions (markdown-ready)
        1 validation plot (for team review)
    """
    # Step 1: Load cleaned data
    df = load_cleaned_data()
    if df is None:
        print(" Exiting: Failed to load cleaned data.")
        return
    
    # Step 2: Run all indicators
    print("\n=== Running All Indicators ===")
    group_query_df, group_desc = group_query_analysis(df)
    transformation_df, transform_desc = data_transformation_analysis(df)
    temporal_df, temporal_desc = temporal_feature_analysis(df)
    spatial_points_df, spatial_centroids_df, spatial_desc = spatial_feature_analysis(df)
    
    # Step 3: Export results to CSV (Member 3 can directly load these into Dash)
    output_files = {
        "1_group_query_top10_collections.csv": group_query_df,
        "2_transformation_rms_quality.csv": transformation_df,
        "3_temporal_monthly_trends.csv": temporal_df,
        "4_spatial_clustering_points.csv": spatial_points_df,
        "4_spatial_clustering_centroids.csv": spatial_centroids_df
    }
    
    for file_name, data in output_files.items():
        data.to_csv(file_name, index=False)
        print(f" Exported: {file_name}")
    
    # Step 4: Export indicator descriptions (markdown-ready for dashboard)
    indicators_descriptions = {
        "Indicator 1 (Group Query)": group_desc,
        "Indicator 2 (Data Transformation)": transform_desc,
        "Indicator 3 (Temporal Analysis)": temporal_desc,
        "Indicator 4 (Spatial Clustering)": spatial_desc
    }
    
    with open("indicator_descriptions.txt", "w", encoding="utf-8") as f:
        for indicator, desc in indicators_descriptions.items():
            f.write(f"## {indicator}\n{desc}\n\n")
    print(f" Exported: indicator_descriptions.txt")
    
    # Step 5: Generate validation plot (fixed all matplotlib compatibility issues)
    plt.rcParams['figure.figsize'] = (16, 12)
    fig, axes = plt.subplots(2, 2)
    
    # Plot 1: Group Query (Accuracy Tier Distribution)
    axes[0,0].bar(
        x=range(len(group_query_df['rms_error_quality'])),
        height=group_query_df['annotation_count'],
        color=['darkgreen', 'lightgreen', 'orange', 'red']  
    )
    axes[0,0].set_xticks(range(len(group_query_df['rms_error_quality'])))
    axes[0,0].set_xticklabels(
        group_query_df['rms_error_quality'],
        rotation=15,  
        ha='right',
        fontsize=10
    )
    axes[0,0].set_ylim(0, group_query_df['annotation_count'].max() * 1.1)
    axes[0,0].set_title("Annotation Count by Georeferencing Accuracy", fontsize=12, fontweight='bold')
    axes[0,0].set_xlabel("Accuracy Tier")
    axes[0,0].set_ylabel("Annotation Count")
    
    # Plot 2: Data Transformation (Pie Chart) - Fixed color list format
    colors = ['skyblue', 'orange', 'green', 'red'] 
    axes[0,1].pie(transformation_df['count'], labels=transformation_df['rms_error_quality'], autopct='%1.1f%%', colors=colors)
    axes[0,1].set_title("Georeferencing Accuracy Distribution", fontsize=12, fontweight='bold')
    
    # Plot 3: Temporal Analysis (Line Chart) - Fixed category type error
    temporal_historical = temporal_df.dropna(subset=['annotation_count']).copy()
    temporal_forecast = temporal_df.dropna(subset=['forecasted_annotations']).copy()
    
    temporal_historical['year_month'] = temporal_historical['year_month'].astype(str)
    if not temporal_forecast.empty:
        temporal_forecast['year_month'] = temporal_forecast['year_month'].astype(str)
        axes[1,0].plot(temporal_forecast['year_month'], temporal_forecast['forecasted_annotations'], marker='s', color='red', linestyle='--', label='Forecasted')
    
    axes[1,0].plot(temporal_historical['year_month'], temporal_historical['annotation_count'], marker='o', color='darkblue', label='Historical')
    axes[1,0].set_title("Monthly Annotation Trends (2014)", fontsize=12, fontweight='bold')
    axes[1,0].set_xlabel("Year-Month")
    axes[1,0].set_ylabel("Annotation Count")
    axes[1,0].tick_params(axis='x', rotation=45)
    axes[1,0].legend()
    
    # Plot 4: Spatial Clustering (Scatter Plot) 
    scatter = axes[1,1].scatter(
        spatial_points_df['center_longitude'], 
        spatial_points_df['center_latitude'],
        c=spatial_points_df['cluster_label'],
        cmap='viridis',
        alpha=0.6,
        s=50
    )
    axes[1,1].scatter(
        spatial_centroids_df['centroid_longitude'],
        spatial_centroids_df['centroid_latitude'],
        marker='X',
        s=200,
        c='red',
        edgecolor='black',
        label='Centroids'
    )
    axes[1,1].set_title("Annotation Hotspots (K-Means Clustering)", fontsize=12, fontweight='bold')
    axes[1,1].set_xlabel("Longitude")
    axes[1,1].set_ylabel("Latitude")
    axes[1,1].legend()
    plt.colorbar(scatter, ax=axes[1,1], label='Cluster Label')
    
    # Adjust layout and save plot
    plt.tight_layout()
    plt.savefig("validation_plots.png", dpi=300, bbox_inches='tight')
    plt.close()
    print(f" Exported: validation_plots.png (team review)")
    
    print("\nAll indicators completed successfully!")
    print("Share the following files with Member 3:")
    print("- All 5 CSV files")
    print("- indicator_descriptions.txt")
    print("- validation_plots.png (optional)")

if __name__ == "__main__":
    main()
```

```python
import dash
from dash import dcc, html
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
```

```python
def load_data():
    """
    Name: load_data
    Input: Reads fixed CSV filenames generated by jinxin
    Output: A dictionary containing pandas DataFrames for each indicator
    Description: Loads the CSV files produced by the analysis pipeline.
    """
    try:
        data = {
            "group": pd.read_csv("1_group_query_top10_collections.csv"),
            "transform": pd.read_csv("2_transformation_rms_quality.csv"),
            "temporal": pd.read_csv("3_temporal_monthly_trends.csv"),
            "spatial_points": pd.read_csv("4_spatial_clustering_points.csv"),
            "spatial_centroids": pd.read_csv("4_spatial_clustering_centroids.csv")
        }
        return data
    except FileNotFoundError as e:
        print(f"Error loading data: {e}")
        return None
```

### Visualization Functions Corresponding to Analysis Elements

The following four functions are designed to generate the charts for the dashboard, directly corresponding to the four key analysis elements derived in the previous steps:

1.  **`create_quality_bar_chart`**: Generates a **Bar Chart** corresponding to **Element 1: Group Query (Accuracy Count)**.
2.  **`create_quality_pie_chart`**: Generates a **Pie Chart** corresponding to **Element 2: Data Transformation (Quality Distribution)**.
3.  **`create_temporal_line_chart`**: Generates a **Line Chart** corresponding to **Element 3: Temporal Analysis (Monthly Trends & Forecast)**.
4.  **`create_spatial_map`**: Generates a **Map Scatter** corresponding to **Element 4: Spatial Clustering (Geographic Hotspots)**.

```python
def create_quality_bar_chart(df):
    """
    Name: create_quality_bar_chart
    Input: df (pd.DataFrame) - Data containing 'rms_error_quality' and 'annotation_count'
    Output: fig (plotly.graph_objects.Figure)
    Description: Creates a bar chart showing the number of annotations per quality tier.
    """
    fig = px.bar(
        df, 
        x='rms_error_quality', 
        y='annotation_count',
        title='Distribution of Georeferencing Accuracy',
        labels={'rms_error_quality': 'Accuracy Tier', 'annotation_count': 'Number of Maps'},
        color='rms_error_quality',
        color_discrete_sequence=px.colors.qualitative.Prism
    )
    fig.update_layout(showlegend=False)
    return fig
```

```python
def create_quality_pie_chart(df):
    """
    Name: create_quality_pie_chart
    Input: df (pd.DataFrame) - Data containing 'rms_error_quality' and 'count'
    Output: fig (plotly.graph_objects.Figure)
    Description: Creates a pie chart showing the percentage share of each quality tier.
    """
    fig = px.pie(
        df, 
        names='rms_error_quality', 
        values='count',
        title='Proportion of Accuracy Categories',
        hole=0.4
    )
    fig.update_traces(textinfo='percent+label')
    return fig
```

```python
def create_temporal_line_chart(df):
    """
    Name: create_temporal_line_chart
    Input: df (pd.DataFrame) - Data containing 'year_month', 'annotation_count', 'forecasted_annotations'
    Output: fig (plotly.graph_objects.Figure)
    Description: Creates a dual-line chart showing historical data and ARIMA forecasts.
    """
    fig = go.Figure()

    # Historical Data Line
    fig.add_trace(go.Scatter(
        x=df['year_month'], 
        y=df['annotation_count'],
        mode='lines+markers',
        name='Historical Data',
        line=dict(color='blue')
    ))

    # Forecast Data Line (Only plot where we have forecast values)
    forecast_data = df.dropna(subset=['forecasted_annotations'])
    if not forecast_data.empty:
        fig.add_trace(go.Scatter(
            x=forecast_data['year_month'], 
            y=forecast_data['forecasted_annotations'],
            mode='lines+markers',
            name='Forecast (ARIMA)',
            line=dict(color='red', dash='dash')
        ))

    fig.update_layout(
        title='Monthly Annotation Trends & Short-term Forecast',
        xaxis_title='Month',
        yaxis_title='Number of Annotations',
        hovermode="x unified"
    )
    return fig
```

```python
def create_spatial_map(points_df, centroids_df):
    """
    Name: create_spatial_map
    Input: points_df, centroids_df
    Output: fig (plotly.graph_objects.Figure)
    Description: 
        Creates a map scatter plot. 
        [BONUS]: Integrates an external dataset of 'Major Cultural Capitals' to interpret the hotspots.
    """
    # Base Map with Points
    fig = px.scatter_map(
        points_df, 
        lat="center_latitude", 
        lon="center_longitude", 
        color="cluster_label",
        size_max=10, 
        zoom=1,
        opacity=0.5,
        title="Geographic Hotspots vs. Major Cultural Capitals (External Data)"
    )
    
    # Cluster Centers
    fig.add_trace(go.Scattermap(
        lat=centroids_df['centroid_latitude'],
        lon=centroids_df['centroid_longitude'],
        mode='markers',
        marker=go.scattermap.Marker(size=18, color='red', symbol='circle'),
        text=centroids_df['cluster_name'],
        name='Cluster Centers (Analysis)'
    ))

    #External Dataset
    external_data = pd.DataFrame({
        'City': ['London (British Library)', 'Paris', 'New York', 'Rome', 'Berlin'],
        'lat': [51.5074, 48.8566, 40.7128, 41.9028, 52.5200],
        'lon': [-0.1278, 2.3522, -74.0060, 12.4964, 13.4050]
    })

    # Adding External Data Points
    fig.add_trace(go.Scattermap(
        lat=external_data['lat'],
        lon=external_data['lon'],
        mode='markers+text',
        marker=go.scattermap.Marker(size=12, color='black', symbol='star'),
        text=external_data['City'],
        textposition="top right",
        name='External Data: Major Cities'
    ))
    fig.update_layout(map_style="open-street-map")
    fig.update_layout(margin={"r":0,"t":40,"l":0,"b":0})
    
    return fig
```
